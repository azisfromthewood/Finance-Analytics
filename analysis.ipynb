```python
#turns the apple stock dataset into a csv file for later use in case apple api doesnt work later on
#df variable which reads the csv file 

import pandas as pd 
aapl.to_csv('appleset.csv', encoding = 'utf-8')

#now takes in the csv file 
df = pd.read_csv('appleset.csv', header = 0 , index_col = 'Date', parse_dates = True)

```

```python
df.head()
#creates a new column with difference between open and close prices
df['diff']= df.Open-df.Close

df.head()
```

```python

#imports matplotlib to plot the graph
import matplotlib.pyplot as plt
#plots the closing timeline over the years
df['Close'].plot(grid = True)
```

```python
#plots the fluctuation in differences
df['diff'].plot(grid = True)
```

```python
import numpy as np 

#adj close to daily close assignment
daily_close = df['Adj Close']

#checking out last closed figures
daily_close.tail()
```

```python
#calculates percent change 
daily_pct_change = daily_close.pct_change()

#replace na values with 0 
daily_pct_change.fillna(0,inplace = True)
print(daily_pct_change)
```

```python
#daily log returns
daily_log_returns = np.log(daily_close.pct_change()+1)
print(daily_log_returns)

daily_log_returns.plot(grid = True)
```
